### 마스터 데이터를 결합(조인)

```python
join_data = pd.merge(join_data, customer_master, on="customer_id", how="left")
join_data = pd.merge(join_data, item_master, on="item_id", how="left")
join_data.head(10)
```

### 필요한 데이터만 뽑기

```python
# 기존 컬럼 데이터를 사용해 새로운 컬럼 제작
join_data["price"] = join_data["quantity"] * join_data["item_price"]
# 필요한 칼럼만 정해서 출력
join_data[["quantity", "item_price", "price"]].head()
```

### 데이터 검산

```python
print(join_data["price"].sum()) # quantity와 item_price로 만든 price
print(transaction["price"].sum())
```

### 월별로 데이터 집계

```python
join_data.dtypes
```

```python
join_data.info()
```

```python
# 시계열 분석에 쓰이는 to_datetime
join_data["payment_date"] = pd.to_datetime(join_data["payment_date"])
join_data["payment_month"] = join_data["payment_date"].dt.strftime("%Y%m")
join_data[["payment_date", "payment_month"]].head()
```

```python
type(join_data["payment_date"].dtypes)
```

```python
join_data.groupby("payment_month")['price'].sum() # 사용방법 변경됨
```

```python
join_data.groupby("payment_month")[['quantity','price']].sum()
```

### 월별, 상품별 데이터를 집계

```python
join_data.groupby(["payment_month","item_name"])[["price", "quantity"]].sum()
```

```python
# 피벗 테이블
pd.pivot_table(join_data, index='item_name', # 행
               columns='payment_month', # 열
               values=['price', 'quantity'], # 값 
               aggfunc='sum') # 합계
```

### 상품별 매출 추이

```python
graph_data = pd.pivot_table(join_data, index='payment_month', 
                            columns='item_name', 
                            values='price', 
                            aggfunc='sum')
graph_data.head()
```

```python
import matplotlib.pyplot as plt
# %matplotlib inline
plt.plot(list(graph_data.index), graph_data["PC-A"], 'o--', label='PC-A')
plt.plot(list(graph_data.index), graph_data["PC-B"], 'o--', label='PC-B')
plt.plot(list(graph_data.index), graph_data["PC-C"], 'o--', label='PC-C')
plt.plot(list(graph_data.index), graph_data["PC-D"], 'o--', label='PC-D')
plt.plot(list(graph_data.index), graph_data["PC-E"], 'o--', label='PC-E')
plt.legend()  
plt.show()
```
