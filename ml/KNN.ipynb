```python
import numpy as np
```

```python
bream_length_array = np.array(bream_length)
bream_weight_array = np.array(bream_weight)
```

```python
length_mean = bream_length_array.mean()
weight_mean = bream_weight_array.mean()
print(length_mean, weight_mean)
```

```python
n = len(bream_length_array)
```

```python
covariance = np.sum((bream_length_array - length_mean) * (bream_weight_array - weight_mean)) / n
print(covariance)
```

```python
std_dev_length = np.sqrt(np.sum((bream_length_array - length_mean) ** 2) / n)
std_dev_weight = np.sqrt(np.sum((bream_weight_array - weight_mean) ** 2) / n)
print(std_dev_length, std_dev_weight)
```

```python
print(f"{covariance / (std_dev_length * std_dev_weight):.10f}")
```
