```python
# 3-1 확인문제
# 1. 2 
# 2. 
knr10 = KNeighborsRegressor()
x = np.arange(5, 45).reshape(-1, 1)

for n in [1, 5, 10]:
    knr10.n_neighbors = n
    knr10.fit(X_train2d, y_train)

    prediction = knr10.predict(x)

    plt.scatter(X_train2d, y_train)
    plt.plot(x, prediction)
    plt.title(f'n_neighbors = {n}')
    plt.xlabel('lenght')
    plt.ylabel('weight')
    plt.show()
```
