The color perception and color model theories are pretty detailed and date back to the 19th century. Our understanding of how colors are perceived and created is quite matured. Given this fact, the concept of **color spaces** is an important one to understand. In this notebook, we shall look at three different color spaces and how they help us generate colored images.

```python
import os
import numpy as np
from skimage.color import rgb2lab,lab2rgb, convert_colorspace
from skimage.io import imsave, imread, imshow


import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def extract_rgb_channel(image, color):
    if color not in (0, 1, 2): return image
    c = image[..., color]
    z = np.zeros_like(c)
    return np.array([(c, z, z), (z, c, z), (z, z, c)][color]).transpose(1,2,0)

def extract_lab_channel(image,dimension):
    lab_channel = np.zeros(image.shape)
    if dimension != 0:
        lab_channel[:,:,0] = 80
    lab_channel[:,:,dimension] = image[:,:,dimension]
    return lab2rgb(lab_channel)

def extract_yuv_channel(image,dimension):
    yuv_channel = np.zeros(image.shape)
    if dimension != 0:
        yuv_channel[:,:,0] = 80
    yuv_channel[:,:,dimension] = image[:,:,dimension]
    return convert_colorspace(yuv_channel,'YUV','RGB')
```

```python
def grid_display(image_directory=None,no_of_columns=2, figsize=(10,10)):
    fig = plt.figure(figsize=figsize)
    column = 0
    list_of_images = os.listdir(image_directory)
    for i in range(len(list_of_images)):
        column += 1
        #  check for end of column and create a new figure
        if column == no_of_columns+1:
            fig = plt.figure(figsize=figsize)
            column = 1
        fig.add_subplot(1, no_of_columns, column)
        plt.imshow(imread(image_directory+list_of_images[i]))
        plt.axis('off')
        plt.title(list_of_images[i])
```

```python
def render_color_spaces(image_file_path=None,
                        target_color_spaces=['RGB','LAB'],
                        figsize=(10,10)):
    
    plt.figure(figsize=figsize)
    
    row = 0
    for i in target_color_spaces:
        plt.subplot2grid((4*len(target_color_spaces),16), (row, 0), rowspan=4, colspan=4)
        img_array = imread(image_file_path)[:,:,:3]
        plt.imshow(img_array)
        plt.axis('off')
        plt.title(image_file_path.split('/')[1])
        
        if i == 'LAB':
            target_img_array = rgb2lab(img_array)
        else:
            target_img_array = convert_colorspace(img_array,'RGB',i)
        for col in [0,1,2]:
            plt.subplot2grid((4*len(target_color_spaces),16), 
                                  (row, (col+1)*4),rowspan=4, colspan=4)
            if i == 'LAB':
                plt.imshow(extract_lab_channel(target_img_array, col))
            elif i=='YUV':
                plt.imshow(extract_yuv_channel(target_img_array, col))
            else:
                plt.imshow(extract_rgb_channel(target_img_array, col))
            plt.title(i[col])
            plt.axis('off')
        row += 4
        #plt.show()
```

## Color Spaces

In simple terms, color spaces are specific organization of colors. They are utilized to reproduce colors using analog and digital representations. The theory goes back to the early days of Thomas Young when he postulated the existence of three types of photoreceptors. Since then different color spaces have come up to reproduce colors, both in print and digital media/screens. Following are a few color spaces: 

* **RGB** : This is a very widely used color space/color model for screens. It is an additive model which uses Red, Green and Blue colors in different combinations to produce a very wide array of colors. It is closesly associated with human perception of colors.


* **YUV** : The __Y__ stands for luma while __U__ and __V__ channels stand for chrominance. This encoding scheme is utilized widely in video systems to map human color perception. The UV channels mainly assist in determining relative amounts red and blue colors. This scheme was widely used due to its capability of using lesser bandwidth and being less prone to transmission errors.


* **LAB**: This device independent color space reference was developed by  International Commission on Illumination. The __L__ channel represents lightness of color (0 is black while 100 is diffuse white), __a__ represents positions between green and magenta while __b__ represens positions between  blue and yellow. 


Below are a few sample images represented in RGB, YUV and LAB channels for a better understanding
