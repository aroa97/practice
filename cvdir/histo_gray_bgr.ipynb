```python
import cv2
import numpy as np
import matplotlib.pylab as plt
```

```python
#--① 이미지 그레이 스케일로 읽기 및 출력
img = cv2.imread('./img/mountain.jpg', cv2.IMREAD_GRAYSCALE)
cv2.imshow('img', img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img
```

```python
img.shape
```

```python
#--② 히스토그램 계산 및 그리기
hist = cv2.calcHist([img], [0], None, [256], [0,255])
plt.plot(hist)
plt.show()
```

```python
hist
```

```python
print("hist.shape:", hist.shape)  #--③ 히스토그램의 shape (256,1)
print("hist.sum():", hist.sum(), "img.shape:",img.shape) #--④ 히스토그램 총 합계와 이미지의 크기
```

```python
import cv2
import numpy as np
import matplotlib.pylab as plt
```

```python
#--① 이미지 그레이 스케일로 읽기 및 출력
img = cv2.imread('./img/cat01.jpg', cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img, (600,600))
cv2.imshow('img', img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#--② 히스토그램 계산 및 그리기
hist = cv2.calcHist([img], [0], None, [256], [0,255])
plt.plot(hist)
plt.show()
```

```python
#--① 이미지 그레이 스케일로 읽기 및 출력
img = cv2.imread('./img/cat01.jpg')
img = cv2.resize(img, (600,600))
```

```python
channels = cv2.split(img)
colors = ('b', 'g', 'r')
for (ch, color) in zip (channels, colors):
    hist = cv2.calcHist([ch], [0], None, [256], [0, 255])
    plt.plot(hist, color = color)
plt.show()
```
