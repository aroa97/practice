```python
import cv2
import numpy as np
import matplotlib.pylab as plt
```

```python
img = cv2.imread('./img/parrot.jpg')
cv2.imshow('img', img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
channels = cv2.split(img_hsv)
```

```python
channels
```

```python
colors = ('h', 's', 'v')
for (ch, color) in zip (channels, colors):
    hist = cv2.calcHist([ch], [0], None, [256], [0, 255])
    plt.plot(hist)
plt.show()
```
