```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import pandas as pd
```

```python
df1 = pd.read_csv('C:/ex/mldl/data/sonar3_copy.csv', header=None)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# 기본값 훈련셋  테스트셋
#        0.75  :  0.25
```

```python
X = df1.iloc[:, :-1].values
y = df1.iloc[:, -1].values
```

```python
train_test_split(X, y, test_size=0.3, shuffle=True)
```

```python
208 * 7 / 10
```

```python
208 * 3 / 10
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
model = Sequential()
model.add(Dense(36, input_dim=60, activation='relu'))
model.add(Dense(18, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=200)
```

```python
# 실제 모델의 정확도는 테스트용으로 확인해야함
```

```python
model.evaluate(X_test, y_test)
```

```python
score = model.evaluate(X_test, y_test)
```

```python
# 테스트 셋에서의 정확도를 가지고 판단해야함 71.42%
```

```python
# 교재 방식
```

```python
model2 = Sequential()
model2.add(Dense(24, input_dim=60, activation='relu'))
model2.add(Dense(10, activation='relu'))
model2.add(Dense(1, activation='sigmoid'))
model2.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model2.fit(X_train, y_train, epochs=200, batch_size=10)
```

```python
score2 = model2.evaluate(X_test, y_test)
```
