```python
x
```

```python
# 1. 상수 함수의 미분
f1 = 5
```

```python
sp.diff(f1, x)
```

```python
# 거듭제곱(역) 함수
```

$y = x^4$

```python
f2 = x**4
sp.diff(f2, x)
```

$ y = x^2 - 5x $

```python
f3 = x**2 - 5 * x
```

```python
sp.diff(f3, x)
```

$y = (3x^2 - 1)(2x^2 + 4x + 2) $

```python
f4 = (3 * x ** 2 - 1) * (2 * x ** 2 + 4 * x + 2)
```

```python
sp.diff(f4, x)
```

$ y = \frac{2x^2 + 3x} {x^2 + 1} $

```python
f5 = (2 * x ** 2 + 3 * x) / (x ** 2 + 1)
```

```python
sp.diff(f5, x)
```

$ (3x^2 + 2)^2 $

```python
f6 = (3 * x ** 2 + 2) ** 2
```

```python
sp.diff(f6, x)
```

$\frac{1}{(2x^2 + 3x - 1)^3}$

```python
f7 = 1 / (2 * x**2 + 3 * x - 1) ** 3
```

```python
sp.diff(f7, x)
```
